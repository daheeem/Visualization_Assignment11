Visualisasi Jalur Transportasi Kereta Api Pada Kota-Kota Besar di Pulau Jawa

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
```

```python
G = nx.DiGraph()

#list kota
kota = ["Jakarta",
        "Surabaya",
        "Bandung",
        "Purwokerto",
        "Tangerang",
        "Semarang",
        "Malang",
        "Tasikmalaya",
        "Cirebon",
        "Pekalongan",
        "Madiun",
        "Yogyakarta",
        "Surakarta",
        "Bogor",
        "Jember",
        ]
G.add_nodes_from(kota)

#rute kereta api
rute = [
    ("Depok", "Bogor"),
    ("Bogor", "Depok"),
    ("Jakarta","Depok"),
    ("Depok", "Jakarta"),
    ("Jakarta","Tangerang"),
    ("Jakarta","Bandung"),

    ("Bandung", "Bogor"),
    ("Bandung", "Tasikmalaya"),
    ("Bandung", "Cirebon"),

    ("Cirebon", "Pekalongan"),

    ("Tasikmalaya", "Purwokerto"),
    ("Purwokerto", "Yogyakarta"),

    ("Semarang", "Pekalongan"),
    ("Semarang", "Surabaya"),
    ("Semarang", "Surakarta"),

    ("Surakarta", "Yogyakarta"),
    ("Surakarta", "Madiun"),
    ("Madiun", "Malang"),
    ("Madiun", "Surabaya"),
    ("Surabaya", "Jember"),
    ("Surabaya", "Malang"),
    ("Malang", "Jember"),


]
G.add_edges_from(rute)

nx.draw(G,font_size=10,node_size=800,node_color='blue', with_labels=True)
plt.figure(figsize=(10, 5))
plt.show()
```

```python
G = nx.Graph()

#list kota
kota = ["Jakarta",
        "Surabaya",
        "Bandung",
        "Purwokerto",
        "Tangerang",
        "Semarang",
        "Malang",
        "Tasikmalaya",
        "Cirebon",
        "Pekalongan",
        "Madiun",
        "Yogyakarta",
        "Surakarta",
        "Bogor",
        "Jember",
        ]
G.add_nodes_from(kota)

#rute kereta api
rute = [
    ("Depok", "Bogor"),
    ("Jakarta","Depok"),
    ("Jakarta","Tangerang"),
    ("Jakarta","Bandung"),

    ("Bandung", "Bogor"),
    ("Bandung", "Tasikmalaya"),
    ("Bandung", "Cirebon"),

    ("Cirebon", "Pekalongan"),

    ("Tasikmalaya", "Purwokerto"),
    ("Purwokerto", "Yogyakarta"),

    ("Semarang", "Pekalongan"),
    ("Semarang", "Surabaya"),
    ("Semarang", "Surakarta"),

    ("Surakarta", "Yogyakarta"),
    ("Surakarta", "Madiun"),
    ("Madiun", "Malang"),
    ("Madiun", "Surabaya"),
    ("Surabaya", "Jember"),
    ("Surabaya", "Malang"),
    ("Malang", "Jember"),


]
G.add_edges_from(rute)

nx.draw(G,font_size=10,node_size=800,node_color='blue', with_labels=True)
plt.figure(figsize=(10, 5))
plt.show()
```

```python
nx.degree_centrality(G)
```

```python
nx.betweenness_centrality(G)
```

```python
degree_centrality = nx.degree_centrality(G)
node_size = [2000 * degree_centrality[node] for node in G.nodes()]

pos = nx.spring_layout(G, seed=42)
nx.draw_networkx(G, pos, with_labels=True, node_size=node_size, node_color='blue', font_size=10)
```
